```python
from scipy import interpolate
from scipy.optimize import curve_fit
import numpy as np 
import matplotlib.pyplot as plt

class seriesejercicio():
    tiempo = []
    columnas = []
    rutadelarchivo = ''
    titulos = []

    def leerdatos(self,rutadelarchivo):
        #apertura de datos
        fichero = open(rutadelarchivo, 'r')
        #volcado de datos
        self.titulos = fichero.readline().strip().split(';')
        fichero.seek(0)
        datos = np.genfromtxt(fichero, delimiter=';')
        #cierre
        fichero.close()
        #borrar
        del fichero       

        #tratamiento de datos
        self.tiempo = datos[1:,0]
        self.columnas = datos[1:,1:]

        return datos
    
    def interpolacionColumna(self,columna):
        interp_func = interpolate.interp1d(self.tiempo, columna , kind='linear')
        
        return interp_func
    
    def ajustarFuncion(self, funcion,columna):
        # Realizar el ajuste de la funci√≥n a los datos
        parametros_optimos, covarianza = curve_fit(funcion, self.tiempo, columna)
        #los parametros se devuelven en el mismo orden en el que se definio la funcion
        return parametros_optimos
    
    
```

```python
oscilador = seriesejercicio()
datos = oscilador.leerdatos('oscilador.csv')       
print(type(datos))
print(datos.shape)
print(datos.dtype)
print(oscilador.tiempo)
print(oscilador.columnas[0:,0])
print(type(oscilador.titulos))
print(oscilador.titulos)

def funcion_sinusoidal(x, a, b, c, d):
    return a * np.sin(b * x + c) + d
x = oscilador.tiempo 
y = funcion_sinusoidal(x,a,b,c,d)


numeroGraficas = len(oscilador.titulos) - 1
# Crear subgr√°ficas
fig, axs = plt.subplots(numeroGraficas)

# Trazar cada gr√°fica en su subgr√°fica correspondiente
for i in range(numeroGraficas):
    parametros = oscilador.ajustarFuncion(funcion_sinusoidal,oscilador.columnas[0:,i])
    a ,b , c, d = parametros 
    axs[i].plot(x, y)
    axs[i].set_title(oscilador.titulos[i+1])

# Ajustar los m√°rgenes
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
plt.savefig('grafico oscilador.png')

#periodo

pi = np.pi
periodo = 2*pi/b
```
