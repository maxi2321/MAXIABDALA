```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Definir la arquitectura de la red CNN-LSTM
class CNN(nn.Module):
    def __init__(self, input_shape):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3)  # 32 kernels, kernel size = 3
        self.fc_input_size = self._calculate_conv_output_size(input_shape)
        self.fc = nn.Linear(self.fc_input_size, 64)  # Capa completamente conectada

    def _calculate_conv_output_size(self, input_shape):
        # Calcular el tamaño de la salida de la capa convolucional
        x = torch.randn(1, *input_shape)
        x = self.conv1(x)
        return x.flatten().shape[0]

    def forward(self, x):
        x = self.conv1(x)
        x = torch.relu(x)
        x = x.view(-1, self.fc_input_size)  # Aplanar para la capa completamente conectada
        x = self.fc(x)
        return x

class LSTM(nn.Module):
    def __init__(self):
        super(LSTM, self).__init__()
        self.lstm = nn.LSTM(input_size=64, hidden_size=64, num_layers=1, batch_first=True)

    def forward(self, x):
        output, _ = self.lstm(x)
        return output[:, -1, :]

class CNN_LSTM(nn.Module):
    def __init__(self, input_shape):
        super(CNN_LSTM, self).__init__()
        self.cnn = CNN(input_shape)
        self.lstm = LSTM()
        self.fc1 = nn.Linear(64, 32)
        self.fc2 = nn.Linear(32, 3)  # 3 clases: Auto, Moto, Colectivo

    def forward(self, x):
        x = self.cnn(x)
        x = x.unsqueeze(0)
        x = self.lstm(x)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Definir las dimensiones de los datos de entrenamiento para cada categoría
SAuto_shape = (1, 401, 5)  # Añadir la dimensión de canal para la categoría 'Auto'
SMoto_shape = (1, 551, 5)  # Añadir la dimensión de canal para la categoría 'Moto'
SColectivo_shape = (1, 801, 5)  # Añadir la dimensión de canal para la categoría 'Colectivo'

# Definir los espectrogramas 
SAuto = torch.tensor(SAuto, dtype=torch.float32).unsqueeze(0)
SMoto = torch.tensor(SMoto, dtype=torch.float32).unsqueeze(0)
SColectivo = torch.tensor(SColectivo, dtype=torch.float32).unsqueeze(0)

# Definir los objetivos para cada categoría
targets = torch.tensor([0, 1, 2])  # Categorías 'Auto', 'Moto', 'Colectivo'

# Crear los datos de entrenamiento y los objetivos
S_tensors = [SAuto, SMoto, SColectivo]
targets_tensors = [torch.tensor([0]), torch.tensor([1]), torch.tensor([2])]

# Lista para almacenar las pérdidas en cada época
losses = []

# Usamos la forma del espectrograma más grande como referencia
input_shape = (1, max(SAuto_shape[1], SMoto_shape[1], SColectivo_shape[1]), 5)

# Entrenar y guardar el modelo
model = CNN_LSTM(input_shape)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

# Iteraciones de entrenamiento incremental
for epoch in range(50):  # Número de épocas
    for S_tensor, target in zip(S_tensors, targets_tensors):
        # Ajustar el tamaño del tensor de entrada
        S_tensor_resized = nn.functional.interpolate(S_tensor.unsqueeze(0), size=input_shape[1:]).squeeze(0)
        # Entrenamiento
        optimizer.zero_grad()
        outputs = model(S_tensor_resized)
        loss = criterion(outputs, target)
        loss.backward()
        optimizer.step()
        losses.append(loss.item())

# Guardar el modelo entrenado
torch.save(model.state_dict(), "ClasificadorAudios.pth")

# Graficar la pérdida en función de las épocas
plt.plot(losses)
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.title('Pérdida durante el entrenamiento')
plt.show()


```

## FUNCION DE PREDICCION    

```python
# Función de predicción para un nuevo espectrograma
def predict(model, spectrogram):
    model.eval()
    with torch.no_grad():
        spectrogram_resized = nn.functional.interpolate(spectrogram.unsqueeze(0), size=input_shape[1:]).squeeze(0)
        outputs = model(spectrogram_resized)
        _, predicted = torch.max(outputs, 1)
    return predicted.item()
```
